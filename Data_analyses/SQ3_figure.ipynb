```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'SQ3_data.csv'
df_sq3 = pd.read_csv(file_path, sep=';')

df_compared = df_sq3.melt(id_vars='Year', value_vars=['Conventional bike', 'Electric bike', 'Car'], var_name='Vehicle Type', value_name='Fatalities')

plt.figure(figsize=(12, 6))
sns.barplot(data=df_compared, x='Year', y='Fatalities', hue='Vehicle Type')

plt.grid()
plt.legend(title='Vehicle Type', bbox_to_anchor=(1, 1), loc='upper left')
plt.title('Fatal Accidents, Car / Bike')
plt.xlabel('Year')
plt.ylabel('Fatalities')
plt.show()
```
