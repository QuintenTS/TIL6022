```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.animation import FuncAnimation

file_path = r'C:\Users\quint\OneDrive\Documents\OneDrive\Master TIL\Python programming\TIL6022\Data_csv\fatalities_road.csv'
data = pd.read_csv(file_path, delimiter=';')

# Prepare the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Set the number of bars
bar_width = 0.2
x = np.arange(len(data['Year']))

# Initialize the bars
bars1 = ax.bar(x, data['Conventional bike'], width=bar_width, label='Conventional Bike', color='blue')
bars2 = ax.bar(x + bar_width, data['Electric bike'], width=bar_width, label='Electric Bike', color='green', linestyle='dotted')
bars3 = ax.bar(x + 2 * bar_width, data['Total bike'], width=bar_width, label='Total Bike', color='grey', linestyle='dotted')
bars4 = ax.bar(x + 3 * bar_width, data['Car'], width=bar_width, label='Car', color='red')

# Set labels and title
ax.set_ylabel('Number of Fatalities')
ax.set_title('Fatalities by Mode of Transport (2018-2022)')
ax.set_xticks(x + 1.5 * bar_width)  # Set x-ticks to the center of the bars
ax.set_xticklabels(data['Year'])
ax.set_ylim(0, data[['Conventional bike', 'Electric bike', 'Total bike', 'Car']].max().max() + 10)  # Set y-axis limit

# Add legend
ax.legend()

# Animation function
def update(year):
    for bar, value in zip(bars1, data['Conventional bike'].iloc[year]):
        bar.set_height(value)
    for bar, value in zip(bars2, data['Electric bike'].iloc[year]):
        bar.set_height(value)
    for bar, value in zip(bars3, data['Total bike'].iloc[year]):
        bar.set_height(value)
    for bar, value in zip(bars4, data['Car'].iloc[year]):
        bar.set_height(value)
    ax.set_title(f'Fatalities by Mode of Transport in {data["Year"].iloc[year]}')


# Show the animation
plt.show()
```
